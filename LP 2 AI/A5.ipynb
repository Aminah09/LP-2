Write a program for the Tic-Tac-Toe game using mini-max algorithm

```python
import math

# the game board
board = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']

# the players
player_human = 'X'
player_computer = 'O'

# the mini-max algorithm
def minimax(depth, is_maximizing):
    # check for win or draw
    winner = check_winner()
    if winner == player_computer:
        return 1
    elif winner == player_human:
        return -1
    elif winner == 'tie':
        return 0
    
    # if it's the computer's turn
    if is_maximizing:
        best_score = -math.inf
        for i in range(len(board)):
            if board[i] == ' ':
                board[i] = player_computer
                score = minimax(depth+1, False)
                board[i] = ' '
                best_score = max(score, best_score)
        return best_score
    
    # if it's the human's turn
    else:
        best_score = math.inf
        for i in range(len(board)):
            if board[i] == ' ':
                board[i] = player_human
                score = minimax(depth+1, True)
                board[i] = ' '
                best_score = min(score, best_score)
        return best_score

# check for winner or draw
def check_winner():
    for i in range(0, 9, 3):
        if board[i] == board[i+1] == board[i+2] != ' ':
            return board[i]
    for i in range(3):
        if board[i] == board[i+3] == board[i+6] != ' ':
            return board[i]
    if board[0] == board[4] == board[8] != ' ':
        return board[0]
    if board[2] == board[4] == board[6] != ' ':
        return board[2]
    if ' ' not in board:
        return 'tie'
    return None

# print the board
def print_board():
    print('-------------')
    print('| ' + board[0] + ' | ' + board[1] + ' | ' + board[2] + ' |')
    print('-------------')
    print('| ' + board[3] + ' | ' + board[4] + ' | ' + board[5] + ' |')
    print('-------------')
    print('| ' + board[6] + ' | ' + board[7] + ' | ' + board[8] + ' |')
    print('-------------')

# main game loop
def main():
    print('Tic-Tac-Toe')
    print_board()
    while True:
        # human's turn
        while True:
            index = int(input('Your turn: (0-8) '))
            if board[index] == ' ':
                board[index] = player_human
                break
            else:
                print('That spot is already taken.')
        print_board()
        winner = check_winner()
        if winner is not None:
            break
        
        # computer's turn
        best_score = -math.inf
        best_move = None
        for i in range(len(board)):
            if board[i] == ' ':
                board[i] = player_computer
                score = minimax(0, False)
                board[i] = ' '
                if score > best_score:
                    best_score = score
                    best_move = i
        board[best_move] = player_computer
        print('Computer\'s turn:')
        print_board()
        winner = check_winner()
        if winner is not None:
            break
    
    # game is over
    if winner == player_computer:
        print('Computer wins!')
    elif winner == player_human:
        print('You win!')
    else:
        print('It\'s a tie!')
    print_board()

# start the game
if __name__ == '__main__':
    main()
```
