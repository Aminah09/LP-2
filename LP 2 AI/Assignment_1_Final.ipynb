```python


import sys

Rows = int(input("Enter the number of Rows : "))
Columns = int(input( "Enter the number of Columns : "))

dist_matrix = []

print("\nEnter the values row-wise : ")

for i in range(Rows): 
    a = list(map(int, input().split())) 
    dist_matrix.append(a)
print("\n\nThe distance matrix is as below : \n")
for i in range(Rows):
    for j in range(Columns):
        print(dist_matrix[i][j], end = " ")
    print()

def nearest_neighbor(curr, unvisited, dist_matrix):
    nearest = sys.maxsize
    neighbor = None
    
    for city in unvisited:
        if dist_matrix[curr][city] < nearest:
            nearest = dist_matrix[curr][city]
            neighbor = city
    return neighbor, nearest

def tsp_nn(dist_matrix):
    n = len(dist_matrix)
    Tourlist = []
    Costlist = []
    
    for j in range(0,n):
        tour=[j]*(n+1)
        unvisited=set(range(0, n))
        unvisited.remove(j)
        curr_city=j
        for i in range(1,n):
            next_city,dist=nearest_neighbor(curr_city,unvisited,dist_matrix)
            tour[i]=next_city
            curr_city=next_city
            unvisited.remove(next_city)
            
        tour[0]=j
            
        cost=sum(dist_matrix[tour[i]][tour[i+1]] for i in range(n-1))
        cost+=dist_matrix[tour[n-1]][tour[0]]
        
        print("\n\nFor travelling starting from city : ", j)
        print("\nTour", tour)
        print("Total Cost ", cost)
        
        Tourlist.append(tour)
        Costlist.append(cost)
        
    print("\nThe minimum cost is:",min(Costlist))
    index=Costlist.index(min(Costlist))
    print("The optimised path is:",Tourlist[index])

tsp_nn(dist_matrix)
  
```

<b> 
The code starts by importing the sys module, which is used to set the maximum value for the initial nearest neighbor distance.

The user is prompted to enter the number of rows and columns for the distance matrix.

An empty list called dist_matrix is created to store the distance values.

The user is then prompted to enter the values for the distance matrix row-wise. The input is split, converted to integers, and appended to the dist_matrix list.

Next, the code prints the entered distance matrix.

The nearest_neighbor function is defined, which takes three parameters: curr (the current city), unvisited (a set of unvisited cities), and dist_matrix (the distance matrix). This function finds the nearest neighbor of the current city by iterating over the unvisited cities and comparing the distances in the distance matrix. It returns the nearest neighbor and the corresponding distance.

The tsp_nn function is defined, which takes the distance matrix as input. It initializes empty lists to store the tour paths and their corresponding costs.

The function then iterates over all cities (from 0 to n-1) as the starting city for the tour.

For each starting city, a tour list is created, initialized with the starting city index. An unvisited set is also created, containing all cities except the starting city.

The algorithm iteratively selects the nearest neighbor from the current city and updates the tour and unvisited set until all cities have been visited.

After completing the tour, the function calculates the cost of the tour by summing up the distances between consecutive cities and adding the distance from the last city back to the starting city.

The tour path and cost are then printed for each starting city.

The tour and cost for each starting city are appended to the Tourlist and Costlist respectively.

Finally, the function determines the tour with the minimum cost, prints it, and displays the minimum cost and the optimized path.

The tsp_nn function is called with the dist_matrix as the argument to execute the algorithm.</b>
